```python
import pandas as pd
```

```python
import random
```

```python
lst = ['robot'] * 10
```

```python
lst += ['human'] * 10
```

```python
random.shuffle(lst)
lst
```

```python
data = pd.DataFrame({'whoAmI':lst})
data
```

```python
pd.get_dummies(data['whoAmI']) #для проверки
```

```python
datanew=data
datanew
```

```python
nature = data['whoAmI'].unique()
for element in nature:
    datanew[element] = 0
datanew
```

```python
for i, j in enumerate(datanew['whoAmI']):
    datanew.at[i, j] = 1
datanew
```

```python
datanew.drop('whoAmI', axis=1, inplace=True) # вариант без использования get_dummies()
datanew
```
